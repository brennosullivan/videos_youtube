# Brincando com o CHATGPT

```python
import requests
import pandas as pd
inflacao_url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.16121/dados?formato=json'
selic_url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.11/dados?formato=json'
def get_data(url):
    response = requests.get(url)
    data = response.json()
    return data
inflacao_json = get_data(inflacao_url)
selic_json = get_data(selic_url)
inflacao_df = pd.DataFrame(inflacao_json)
selic_df = pd.DataFrame(selic_json)
inflacao_df['data'] = pd.to_datetime(inflacao_df['data'], format='%d/%m/%Y')
inflacao_df.set_index('data', inplace=True)

selic_df['data'] = pd.to_datetime(selic_df['data'], format='%d/%m/%Y')
selic_df.set_index('data', inplace=True)
inflacao_df.rename(columns={'valor': 'inflacao'}, inplace=True)
selic_df.rename(columns={'valor': 'selic'}, inplace=True)
dados_bc = inflacao_df.join(selic_df)

         

```
